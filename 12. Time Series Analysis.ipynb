### Window Function

In window function, we use windows of data to do the analysis. Basically two type of window functions are defined. Rolling window and expending window.
In rolling window, we take the data points and give the combined result at some point of time. We can consider it as moving summation. We can choose different size of rolling window. 
Expending window keep taking the data points. At any point it will give you information considering all the past points. We can consider it as cumulative summation.


Let's create our first rolling window

```python
rolling_window_5 = data.rolling(window = 5)
```

```python
rolling_window_30
```

```python
data['Price'].plot(color = 'gray')
rolling_window_5.mean()['Price'].plot(color = 'blue')
```

```python
#We can change the window size to get smooth results
rolling_window_10 = data.rolling(window = 10)
data['Price'].plot(color = 'gray')
rolling_window_5.mean()['Price'].plot(color = 'blue')
rolling_window_10.mean()['Price'].plot(color = 'red')
```

## Time Series Forecasting

We will take a data and do different transformation on it. Finally we will apply ARIMA model on it to make time series forecasting.
Data is an Air Passenger data. It contains montly passengers of an airline from Jan-49 to Dec-60. Data is available publicly on https://datamarket.com.

```python
air_passengers = pandas.read_csv('.\\data\\air_passengers.csv',parse_dates=['Month'],index_col=0)
```

```python
#Check the initial records
air_passengers.head()
```

```python
air_passengers.plot()
```

```python
#Import the library for Dickey-Fuller Test 
from statsmodels.tsa.stattools import adfuller
```

```python
def test_timeseries_stationary(ts):
    ts = ts['Passenger']
    print('Dickey-Fuller Test:')
    dickey_fuller_test = adfuller(ts, autolag='AIC')
    dickey_fuller_output = pandas.Series(dickey_fuller_test[0:4], index=['Test Statistic','p-value','Number of Lags Used','Observations Used'])
    for key,value in dickey_fuller_test[4].items():
        dickey_fuller_output['Critical Value (%s)'%key] = value
    print(dickey_fuller_output)
```

```python
test_timeseries_stationary(air_passengers)
```

```python
air_passengers_log_value = numpy.log(air_passengers)
```

```python
air_passengers_log_value.plot()
```

```python
air_passengers_log_value_diff = air_passengers_log_value - air_passengers_log_value.shift()
```

```python
air_passengers_log_value_diff.plot()
```

```python
air_passengers_log_value_diff.dropna(inplace=True)
test_timeseries_stationary(air_passengers_log_value_diff)
```

```python
exponential_weighted_average = pandas.ewma(air_passengers_log_value, halflife=12)
plt.plot(air_passengers_log_value)
plt.plot(exponential_weighted_average, color='Green')
```

```python
exponential_weighted_average_diff = air_passengers_log_value - exponential_weighted_average
```

```python
test_timeseries_stationary(exponential_weighted_average_diff)
```

```python
#Import library for ARIMA
from statsmodels.tsa.arima_model import ARIMA
```

```python
#Auto Regressive Model 
ARIMA_model_1 = ARIMA(air_passengers_log_value, order=(2, 1, 0))  
results_AR_1 = ARIMA_model_1.fit(disp=-1)  
plt.plot(air_passengers_log_value_diff)
plt.plot(results_AR_1.fittedvalues, color='Green')
plt.title('AR Model')
```

```python
print('Residual Sum Of Square: %.6f'% sum((results_AR_1.fittedvalues-air_passengers_log_value_diff['Passenger'])**2))
```

```python
#Moving Average Model 
ARIMA_model_2 = ARIMA(air_passengers_log_value, order=(0, 1, 1))  
results_MA_2 = ARIMA_model_2.fit(disp=-1)  
plt.plot(air_passengers_log_value_diff)
plt.plot(results_MA_2.fittedvalues, color='Green')
plt.title('MA Model')
```

```python
print('Residual Sum Of Square: %.6f'% sum((results_MA_2.fittedvalues-air_passengers_log_value_diff['Passenger'])**2))
```

```python
#ARIMA Model (AR + MA)
ARIMA_model_3 = ARIMA(air_passengers_log_value, order=(1, 1, 1))  
results_ARIMA_3 = ARIMA_model_3.fit(disp=-1)  
plt.plot(air_passengers_log_value_diff)
plt.plot(results_ARIMA_3.fittedvalues, color='Green')
plt.title('ARIMA Model (AR + MA)')
```

```python
print('Residual Sum Of Square: %.6f'% sum((results_ARIMA_3.fittedvalues-air_passengers_log_value_diff['Passenger'])**2))
```

```python
ARIMA_diff_values = pandas.Series(results_ARIMA_3.fittedvalues, copy=True)
```

```python
ARIMA_diff_values_cumsum = ARIMA_diff_values.cumsum()
```

```python
ARIMA_log = pandas.Series(float(air_passengers_log_value.ix[0]), index=air_passengers_log_value.index)
ARIMA_log = ARIMA_log.add(ARIMA_diff_values_cumsum,fill_value=0)
```

```python
ARIMA_Result = numpy.exp(ARIMA_log)
```

```python
plt.plot(air_passengers)
plt.plot(ARIMA_Result)
plt.title('ARIMA Model')
```

```python
print('Root Mean Squared Error: %.6f'% numpy.sqrt(sum((ARIMA_Result-air_passengers['Passenger'])**2)/len(air_passengers['Passenger'])))
```
